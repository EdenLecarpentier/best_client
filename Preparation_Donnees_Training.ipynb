```python
application = pd.read_csv(r"application_train.csv")
```

```python
bureau = pd.read_csv(r"bureau.csv")
```

```python
pos = pd.read_csv(r"POS_CASH_balance.csv")
```

```python
card = pd.read_csv(r"credit_card_balance.csv")
```

## bureau.csv

### AMT_CREDIT_SUM_OVERDUE / AMT_CREDIT_MAX_OVERDUE (bureau.csv)

```python
df_1 = bureau[['SK_ID_CURR', 'AMT_CREDIT_SUM_OVERDUE', 'AMT_CREDIT_MAX_OVERDUE']]
df_1 = df_1.fillna(0)
```

```python
summ = df_1.groupby('SK_ID_CURR',as_index = False).agg('mean')
summ
```

## POS_CASH_balance

### Jour de retard de paiement (SK_DPD)

```python
df_2 = pos[['SK_ID_CURR', 'SK_DPD']]
```

```python
summ_1 = df_2.groupby('SK_ID_CURR', as_index=False).agg('mean')
summ_1
```

## credit_card_balance

### Jour de retard carte 'SK_DPD'

```python
card
```

```python
df_3 = card[['SK_ID_CURR', 'SK_DPD']]
df_3 = df_3.rename(columns={'SK_DPD':'SK_DPD_CARD'})
```

```python
summ_2 = df_3.groupby('SK_ID_CURR', as_index=False).agg('mean')
summ_2
```

## application

**Données à encoder :**
- FLAG_OWN_CAR, FLAG_OWN_REALTY, NAME_INCOME_TYPE, NAME_EDUCATION_TYPE, NAME_FAMILY_STATUS, NAME_HOUSING_TYPE, ORGANIZATION_TYPE.

**Données à conserver**
- SK_ID_CURR, TARGET, CNT_CHILDREN, (AMT_ANNUITY/AMT_INCOME_TOTAL), AMT_GOODS_PRICE, DAYS_EMPLOYED, CNT_FAM_MEMBERS, OCCUPATION_TYPE,  REGION_RATING_CLIENT, REGION_RATING_CLIENT_W_CITY, DAYS_LAST_PHONE_CHANGE.

```python
application
```

### Encoding application(train)

```python
one_hot = pd.get_dummies(application[['FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'NAME_INCOME_TYPE',
                                      'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE',
                                      'OCCUPATION_TYPE', 'ORGANIZATION_TYPE']])
```

```python
one_hot
```

### Colonnes à conserver

```python
df_4 = application[['SK_ID_CURR', 'TARGET', 'CNT_CHILDREN', 'AMT_ANNUITY', 'AMT_INCOME_TOTAL', 'AMT_GOODS_PRICE',
                    'DAYS_EMPLOYED', 'CNT_FAM_MEMBERS', 'REGION_RATING_CLIENT', 
                    'REGION_RATING_CLIENT_W_CITY', 'DAYS_LAST_PHONE_CHANGE']]

df_4['INDEBTEDNESS'] = ((df_4['AMT_ANNUITY'])/(df_4['AMT_INCOME_TOTAL']))
df_4
```

### Join encoded columns

```python
df_4 = df_4.join(one_hot)
```

```python
df_4
```

### Joining all columns

```python
joining = pd.concat([df_1, df_2, df_3], axis=0)
```

```python
first_half = joining.groupby('SK_ID_CURR', as_index=False).agg('sum')
first_half
```

### df_final

```python
df_final = pd.merge(df_4, first_half, on="SK_ID_CURR")
df_final
```

```python
df_final.to_csv("final.csv")
```
